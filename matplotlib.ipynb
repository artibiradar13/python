Matplotlib-------for data visualisation

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,52,53,48,54,67,45]
plt.plot(x,y,color='red',linewidth=4,linestyle="dashdot")
plt.title("WEATHER")
plt.xlabel("DAY")
plt.ylabel("temperature")
```

bar chart


```python
np.arange
```

```python
company=['Googlr','amzn','MSFT','FB']
ypos=np.arange(len(company))#convert to int values
plt.xticks(ypos,company)#taking company name
revenue=[90,56,79,35]
profit=[23,35,37,32]

plt.bar(ypos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(ypos+0.2,profit,width=0.4)
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.barh(ypos-0.2,revenue,height=0.4,label='Revenue')
plt.barh(ypos+0.2,profit,height=0.4,label='Profit')
plt.legend()
```

Histograms

```python
blood_sugar=[113,24,35,67,89,56,43,5,143,123,167]



plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,color='pink',)
```

pie chart----------

```python
exp_vals=[1490,700,567,432,187]
exp_labels=['Home Rent','food','car','phone bill','other utilities']



plt.pie(exp_vals,labels=exp_labels,radius=2.5,autopct='%0.0f%%',explode=[0,0.02,0,0,0])
plt.axis("equal")





###save output to another file
plt.savefig("piechart.png",bbox_inches='tight',pad_inches=2,transparent='true')
```
